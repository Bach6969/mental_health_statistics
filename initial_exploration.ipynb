```python
correlation = mh_df["suicide_rate_per100k"].corr(mh_df["psychiatrists_per_million"])
print(correlation)
```

```python
import matplotlib.pyplot as plt

plt.scatter(mh_df["psychiatrists_per_million"], mh_df["suicide_rate_per100k"])
plt.xlabel("Providers per 100k")
plt.ylabel("Suicide Rate")
plt.title("Mental Health Access vs Suicide Rate")
plt.show()
```

```python
plt.bar(mh_df["psychiatrists_per_million"], mh_df["suicide_rate_per100k"])
plt.xlabel("Providers per 100k")
plt.ylabel("Suicide Rate")
plt.title("Mental Health Access vs Suicide Rate")
plt.show()
```

```python
mh_df["provider_level"] = pd.qcut(mh_df["psychiatrists_per_million"], 4, labels=["Low", "Mid-Low", "Mid-High", "High"])

mh_df.groupby("provider_level")["suicide_rate_per100k"].mean()
```
