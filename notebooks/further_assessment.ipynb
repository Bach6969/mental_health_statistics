```python
import matplotlib.pyplot as plt

plt.scatter(df["psychiatrists_per100k"], df["suicide_rate_per100k"])
plt.xlabel("Psychiatrists per 100k")
plt.ylabel("Suicide Rate")
plt.show()
```

```python
df["provider_level"] = pd.qcut(df["psychiatrists_per100k"], 4)

df.groupby("provider_level")["suicide_rate_per100k"].mean()
```

```python
df[[
    "suicide_rate_per100k",
    "psychiatrists_per100k",
    "mh_spend_usd_per_capita",
    "mh_system_score"
]].corr()
```

```python
import statsmodels.api as sm

X = df[[
    "psychiatrists_per100k",
    "mh_spend_usd_per_capita",
    "mh_system_score"
]]
y = df["suicide_rate_per100k"]

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

print(model.summary())
```
