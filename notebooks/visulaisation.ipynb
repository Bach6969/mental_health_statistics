```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/Global_Mental_Health_Crisis_Index_2026.csv")
```

```python
plt.figure()
plt.hist(df['mh_crisis_index'])
plt.title("Distribution of Mental Health Crisis Index")
plt.xlabel("Crisis Index")
plt.ylabel("Frequency")
plt.show()
```

```python
grouped = df.groupby('income_group')['mh_crisis_index'].mean()

plt.figure()
grouped.plot(kind='bar')
plt.title("Average Mental Health Crisis Index by Income Group")
plt.xlabel("Income Group")
plt.ylabel("Average Crisis Index")
plt.show()
```

```python
plt.figure()
plt.scatter(df['psychiatrists_per100k'], df['suicide_rate_per100k'])
plt.title("Psychiatrists per 100k vs Suicide Rate")
plt.xlabel("Psychiatrists per 100k")
plt.ylabel("Suicide Rate per 100k")
plt.show()
```

```python
sns.regplot(
    x='psychiatrists_per100k',
    y='suicide_rate_per100k',
    data=df
)
plt.show()
```

```python
sns.scatterplot(
    x='psychiatrists_per100k',
    y='suicide_rate_per100k',
    hue='income_group',
    data=df
)
plt.show()
```
