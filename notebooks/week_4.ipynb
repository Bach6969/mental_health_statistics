```python
df = df[df["suicide_rate_per100k"] >= 0]
```

```python
top_countries = df.sort_values(
    by="suicide_rate_per100k",
    ascending=False
).head(15)

plt.figure(figsize=(10,6))

sns.set_palette("muted")  # calm, readable colors

sns.barplot(
    data=top_countries,
    x="suicide_rate_per100k",
    y="country",
)

plt.title("Top 15 Countries by Suicide Rate (per 100k)")
plt.xlabel("Suicide Rate (per 100k people)")
plt.ylabel("Country")

plt.tight_layout()
plt.show()
```

### What this chart shows
This chart compares suicide rates across countries, highlighting which countries have the highest and lowest rates per 100,000 people.

### Why this chart type
A bar chart is appropriate for comparing values across categories (countries), making it easy to identify differences and rank them visually.

### Design choices
Countries are sorted in descending order to make patterns immediately visible. A simple, consistent color palette is used to reduce visual clutter and keep focus on the data. Clear axis labels and units are included to avoid ambiguity.

### Integrity considerations
The axis starts at zero to avoid exaggerating differences between countries. Only a subset of countries is shown to maintain readability, which means this chart does not represent the full global distribution. Additionally, suicide rates may be underreported in some regions due to stigma or differences in data collection practices.

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="psychiatrists_per100k",
    y="suicide_rate_per100k",
    alpha=0.7
)

sns.regplot(
    data=df,
    x="psychiatrists_per100k",
    y="suicide_rate_per100k",
    scatter=False,
    color="red"
)

plt.title("Psychiatrists vs Suicide Rate")
plt.xlabel("Psychiatrists per 100k")
plt.ylabel("Suicide Rate per 100k")

plt.tight_layout()
plt.show()
```

### What this chart shows
This chart explores the relationship between the number of psychiatrists per 100,000 people and suicide rates across countries.

### Why this chart type
A scatter plot is appropriate for examining the relationship between two continuous variables and identifying potential correlations or patterns.

### Design choices
Points are semi-transparent to reduce overlap and improve readability. A regression line is included to highlight the overall trend in the data. Axis labels clearly indicate units to ensure accurate interpretation.

### Integrity considerations
This visualization shows correlation, not causation. Suicide rates are influenced by many factors beyond psychiatrist availability, including economic conditions, cultural attitudes, and reporting practices. Rows with missing values in either variable were removed, which may exclude countries with limited data and introduce bias.

```python
df_spend = df.dropna(subset=[
    "mh_spend_usd_per_capita",
    "mh_system_score"
])

plt.figure(figsize=(10,6))

sns.set_palette("muted")  # calm, readable colors

sns.scatterplot(
    data=df_spend,
    x="mh_spend_usd_per_capita",
    y="mh_system_score",
    alpha=0.7
)

plt.title("Mental Health Spending vs System Score")
plt.xlabel("Spending (USD per capita)")
plt.ylabel("System Score")

plt.tight_layout()
plt.show()
```

```python
### What this chart shows
This chart examines the relationship between mental health spending per capita and overall mental health system performance across countries.

### Why this chart type
A scatter plot allows for direct comparison between two continuous variables and helps reveal whether higher spending is associated with stronger system performance.

### Design choices
A clean, minimal design is used to emphasize the data points. Axis labels include units for clarity. If a log scale is applied to spending, it is used to better represent large differences across countries.

### Integrity considerations
Higher spending does not necessarily cause better system performance, as many structural and policy factors are involved. Spending data may not be directly comparable across countries due to differences in healthcare systems and reporting standards. Any transformations, such as log scaling, are applied to improve readability and are noted for transparency.
```
